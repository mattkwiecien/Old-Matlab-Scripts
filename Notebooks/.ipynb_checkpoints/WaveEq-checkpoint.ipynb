```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import math
import time
from IPython.core.pylabtools import figsize
figsize(8,8)
```

Create a grid of x and y points

```python
c = 1
Lx = 100
Ly = 100
y = np.linspace(0,10,Lx)
x = np.linspace(0,10,Ly)
X, Y = np.meshgrid(x,y)
```

Create a function to calculate a double integral

```python
def double_integral(x,y,f):
    Lx = len(x)-1
    Ly = len(y)-1
    sum = 0
    
    dx = x[0,1] - x[0,0]
    dy = y[1,0] - y[0,0]
    
    for j in range(Ly-1):
        for i in range(Lx-1):
            sum_inner = f[i, j] + f[i, j+1] + f[i+1, j] + f[i+1, j+1]
            sum = sum + (.25 * dx * dy * sum_inner)
            
    return sum
```

Verify our double integral by taking the double integral of Z = 1

```python
y = np.linspace(0,10,Ly)
x = np.linspace(0,10,Lx)
X, Y = np.meshgrid(x,y)
Z, _ = np.meshgrid(np.linspace(1,1,Lx), np.linspace(1,1,Ly))

print(double_integral(X,Y,Z))
```

Create a function which is a gaussian for the initial position

$$ f(x) = $$

```python
def gaussian(X,Y,A,x0,y0,sigX,sigY):
    x_term = ((X-x0)**2)/(2*(sigX**2))
    y_term = ((Y-y0)**2)/(2*(sigY**2))
    return A*np.exp(-1*(x_term+y_term))
```

```python
Z = gaussian(X,Y,5,5,5,1,1)

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.set_xlim3d([0,10])
ax.set_ylim3d([0,10])
ax.set_zlim3d([-5,5])

ax.plot_surface(X,Y,Z,rstride=1, cstride=1, cmap='viridis', edgecolor='none')
```

Now, calculate the fourier coefficients for our general solution to the 2D wave equation

```python
terms = 10
frequency = np.zeros((terms,terms))
A_coeff = np.zeros((terms,terms))
B_coeff = np.zeros((terms,terms))

for m in range(0,terms):
    for n in range(0, terms):
        frequency[m,n] = np.sqrt( (((m+1)**2 * math.pi**2 * c**2)/Lx**2) + (((n+1)**2 * math.pi**2 * c**2)/Ly**2) )
        g = np.sin(X*(m+1)*math.pi/Lx)*np.sin(Y*(n+1)*math.pi/Ly)
        
        A_coeff[m,n] = (4/(Lx*Ly))*double_integral(X,Y,(Z*g))
#         B_coeff=0

total_time = 10
step = 1
frames = total_time / step

Z_anim = np.zeros((Lx, Ly, 10))

for t in np.arange(1, total_time, step):
    Z_anim[:,:,t] = Z
    
    for m in range(0,terms):
        for n in range(0,terms):
            g = np.sin(X*(m+1)*math.pi/Lx)*np.sin(Y*(n+1)*math.pi/Ly)
            Z_anim[:,:,t] += (A_coeff[m,n]*np.cos(frequency[m,n]*t) + B_coeff[m,n]*np.sin(frequency[m,n]*t))*g


def update_figure(frame, zarray, plot):
    plot[0].remove()
    plot[0] = ax.plotsurface(X,Y,zarray[:,:,frame], rstride=1, cstride=1, cmap='viridis', edgecolor='none')

fig=plt.figure()
ax=plt.axes(projection='3d')
ax.set_xlim3d([0,10])
ax.set_ylim3d([0,10])
ax.set_zlim3d([-5,5])

plot = [ax.plot_surface(x, y, Z_anim[:,:,0], rstride=1, cstride=1, cmap='viridis', edgecolor='none')]

line_animation = animation.FuncAnimation(fig, update_figure, 10, fargs=(Z_anim, plot), interval=500, blit=False)
plt.show()
```
